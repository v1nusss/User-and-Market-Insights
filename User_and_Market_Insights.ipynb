```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (replace with your dataset file path)
file_path = '/content/user_and_market_reports.csv'  # Replace with actual file name if different
df = pd.read_csv(file_path)
```

```python
# Display dataset overview
print("Dataset Shape:", df.shape)
print("\nColumn Names:", df.columns)
print("First 5 rows of the dataset:")
df.head()

# Check for null values
print("\nNull values:\n", df.isnull().sum())

# Basic summary statistics
print("\nSummary Statistics:\n", df.describe())

# Data Cleaning: Fill or drop missing values if required
df.fillna(method='ffill', inplace=True)

```

```python
# Visualize insights
## 1. Purchase Frequency Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Purchase_Frequency'], bins=20, kde=True, color='blue')
plt.title('Purchase Frequency Distribution')
plt.xlabel('Purchase Frequency')
plt.ylabel('Count')
plt.show()

```

```python
## 2. Churn Probability by Region
plt.figure(figsize=(12, 6))
sns.boxplot(x='Region', y='Churn_Probability', data=df)
plt.title('Churn Probability by Region')
plt.xlabel('Region')
plt.ylabel('Churn Probability')
plt.show()

```

```python
## 3. Average Order Value by Most Frequent Category
plt.figure(figsize=(12, 6))
sns.barplot(x='Most_Frequent_Category', y='Average_Order_Value', data=df)
plt.title('Average Order Value by Most Frequent Category')
plt.xticks(rotation=45)
plt.ylabel('Average Order Value')
plt.show()

```

```python
## 4. Lifetime Value vs Churn Probability
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Lifetime_Value', y='Churn_Probability', hue='Region', data=df)
plt.title('Lifetime Value vs Churn Probability')
plt.xlabel('Lifetime Value')
plt.ylabel('Churn Probability')
plt.legend(title='Region')
plt.show()

```

```python
# Save cleaned dataset for Tableau
output_path = '/content/cleaned_user_and_market_reports.csv'
df.to_csv(output_path, index=False)
print(f"\nCleaned dataset saved to: {output_path}")
```
